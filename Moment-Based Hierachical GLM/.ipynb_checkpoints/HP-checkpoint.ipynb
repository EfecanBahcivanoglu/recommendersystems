# Gamma Process for User Weights

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma

# Parameters for the gamma process
c = 2.0  # Scaling parameter
H = 1.0  # Base measure

# Assuming a finite number of components for illustration
K = 100  # Number of components

# Draw user weights from the gamma process
theta_u = gamma.rvs(c, scale=1/H, size=K)

# Visualization of user weights
plt.figure(figsize=(10, 6))
plt.bar(range(K), theta_u)
plt.xlabel('Component')
plt.ylabel('Weight')
plt.title('User Weights from Gamma Process')
plt.show()
```

# Beta Process for Item Weights

```python
from scipy.stats import beta

# Parameters for the beta process
alpha = 1.0  # Parameter for the beta distribution

# Draw stick proportions from the beta distribution
omega_uk = beta.rvs(1, alpha, size=K)

# Stick-breaking construction
theta_uk = np.zeros(K)
theta_uk[0] = omega_uk[0]
for k in range(1, K):
    theta_uk[k] = omega_uk[k] * np.prod(1 - omega_uk[:k])

# Visualization of item weights
plt.figure(figsize=(10, 6))
plt.bar(range(K), theta_uk)
plt.xlabel('Component')
plt.ylabel('Weight')
plt.title('Item Weights from Beta Process')
plt.show()
```

# Gamma Distribution for Item Weights

```python
# Parameters for the gamma distribution
a, b = 1.0, 1.0  # Shape and rate parameters

# Draw item weights from the gamma distribution
beta_ik = gamma.rvs(a, scale=1/b, size=K)

# Visualization of item weights
plt.figure(figsize=(10, 6))
plt.bar(range(K), beta_ik)
plt.xlabel('Component')
plt.ylabel('Weight')
plt.title('Item Weights from Gamma Distribution')
plt.show()
```

# Poisson Distributions for Observations 

```python
# Assuming we have arrays of user weights theta_u and item weights beta_ik
# Calculate the rate parameter for Poisson distribution (dot product for each user-item pair)
# Here, we need to ensure that we're working with arrays rather than a single integer.

# For simplicity, let's assume we have one user and multiple items.
# The following code will be adapted for this scenario.

# Example user weights - normally you would have one theta_u per user.
theta_u = np.random.gamma(1.0, 1.0, size=10)  # Assuming 10 latent factors

# Example item weights - one beta_ik per item.
beta_ik = np.random.gamma(1.0, 1.0, size=10)  # Also assuming 10 latent factors

# Calculate the rate parameter for the Poisson distribution
# This would be the dot product of user weights and item weights
# In a real scenario, this should be a matrix multiplication, resulting in a user-item interaction matrix.
# Here, we demonstrate with just a single user.
lambda_ui = np.dot(theta_u, beta_ik)

# Draw observations from the Poisson distribution for each item
y_ui = poisson.rvs(lambda_ui, size=len(beta_ik))

# Visualization of observations
plt.figure(figsize=(10, 6))
plt.bar(range(len(y_ui)), y_ui)
plt.xlabel('Item')
plt.ylabel('Count')
plt.title('User-Item Interactions from Poisson Distribution')
plt.show()
```

```python
import numpy as np
from scipy.stats import gamma, poisson

def generate_data(N, M, a_theta, c_theta, a_beta, b_beta, alpha):
    """
    Generate user-item interaction data based on the provided Bayesian nonparametric model.
    
    Parameters:
    N (int): Number of users.
    M (int): Number of items.
    a_theta, c_theta (float): Parameters for the Gamma distribution for user weights.
    a_beta, b_beta (float): Parameters for the Gamma distribution for item weights.
    alpha (float): Parameter for the Beta distribution for the stick-breaking process.
    
    Returns:
    y_ui (np.array): Generated user-item interaction matrix.
    """
    
    # Step 1: Draw user weights theta_u from a Gamma distribution
    theta_u = gamma.rvs(a_theta, scale=1/c_theta, size=N)
    
    # Step 2: Draw item weights beta_ik from a Gamma distribution
    # We will use a stick-breaking process to ensure the weights are shared among all users
    beta_ik = np.zeros((N, M))
    v = np.zeros(M)
    for k in range(M):
        v[k] = np.random.beta(1, alpha)
        remaining_stick = 1 if k == 0 else remaining_stick * (1 - v[k-1])
        beta_ik[:, k] = remaining_stick * v[k]
    beta_ik *= gamma.rvs(a_beta, scale=1/b_beta, size=M)
    
    # Step 3: Draw Poisson-distributed y_ui for user-item interactions
    y_ui = np.zeros((N, M), dtype=int)
    for u in range(N):
        for i in range(M):
            lambda_ui = theta_u[u] * beta_ik[u, i]  # Poisson rate parameter
            y_ui[u, i] = poisson.rvs(lambda_ui)
    
    return y_ui

# Parameters
N = 10  # Number of users
M = 5   # Number of items
a_theta = 1.0
c_theta = 0.1
a_beta = 1.0
b_beta = 0.1
alpha = 1.0

# Generate the data
data = generate_data(N, M, a_theta, c_theta, a_beta, b_beta, alpha)
print(data)
```

```python
import numpy as np
from scipy.stats import gamma

# Parameters for the Gamma distribution (for simplicity)
a, scale = 2.0, 1.0  # Shape and scale

# Simulate a finite Gamma process
np.random.seed(42)
num_atoms = 100  # A finite number of atoms for demonstration
weights = gamma.rvs(a, scale=scale, size=num_atoms)

# Function for size-biased sampling
def size_biased_ordering(weights):
    ordered_weights = []
    remaining_weights = weights.copy()
    
    for _ in range(len(weights)):
        total_weight = remaining_weights.sum()
        if total_weight == 0:
            break
        
        probabilities = remaining_weights / total_weight
        chosen_index = np.random.choice(len(weights), p=probabilities)
        ordered_weights.append(remaining_weights[chosen_index])
        
        # Reduce the weight of the chosen atom
        remaining_weights[chosen_index] = 0
    
    return np.array(ordered_weights)

# Apply size-biased ordering
size_biased_weights = size_biased_ordering(weights)

# Output for demonstration
print("Original Weights:", weights)
print("Size-Biased Ordered Weights:", size_biased_weights)
```

```python
import numpy as np
from scipy.stats import gamma, poisson

def simulate_gamma_process(shape, scale, num_atoms):
    """Simulate a finite representation of the Gamma process."""
    return gamma.rvs(shape, scale=scale, size=num_atoms)

def size_biased_ordering(weights, truncation_level):
    """Apply size-biased ordering to the weights and truncate."""
    ordered_indices = np.argsort(-weights)  # Sort indices by weight in descending order
    return weights[ordered_indices][:truncation_level]

def generate_item_weights(shape, scale, dimension_item):
    """Generate item weights from a Gamma distribution."""
    return gamma.rvs(shape, scale=scale, size=dimension_item)

def generate_user_item_interactions(user_weights, item_weights, num_users):
    """Generate user-item interactions."""
    interactions = np.zeros((num_users, len(item_weights)), dtype=int)
    for u in range(num_users):
        for i in range(len(item_weights)):
            print(user_weights[u])
            print(item_weights[i])
            rate = user_weights[u] * item_weights[i]
            interactions[u, i] = poisson.rvs(rate)
    return interactions

# Parameters
num_users = 10
num_items = 5
gamma_shape = 2.0
gamma_scale = 1.0
num_atoms = 1000  # Large number for finite representation of Gamma process
truncation_level = 50  # Truncation level for size-biased ordering

# Simulate the Gamma process for user weights
user_weights_process = simulate_gamma_process(gamma_shape, gamma_scale, num_atoms)

# Apply size-biased ordering and truncate
user_weights = np.array([size_biased_ordering(user_weights_process, truncation_level) for _ in range(num_users)])

# Generate item weights
item_weights = generate_item_weights(gamma_shape, gamma_scale, num_items)

print(item_weights)

# Generate user-item interactions
data = generate_user_item_interactions(user_weights, item_weights, num_users)

print("Generated User-Item Interactions Matrix:\n", data)
```
